### Anova-Tests

#### H₀:There are no risk differences across provinces

```python
# Get list of provinces
print(df.groupby('Province')['TotalClaims'].mean().round(2))
```

```python
from scipy import stats

# Separate claim amounts by province
eastern_cape_claims = df[df['Province'] == 'Eastern Cape']['TotalClaims']
free_state_claims = df[df['Province'] == 'Free State']['TotalClaims']
gauteng_claims = df[df['Province'] == 'Gauteng']['TotalClaims']
kwazulu_natal_claims = df[df['Province'] == 'KwaZulu-Natal']['TotalClaims']
limpopo_claims = df[df['Province'] == 'Limpopo']['TotalClaims']
mpumalanga_claims = df[df['Province'] == 'Mpumalanga']['TotalClaims']
north_west_claims = df[df['Province'] == 'North West']['TotalClaims']
northern_cape_claims = df[df['Province'] == 'Northern Cap']['TotalClaims']
western_cape_claims = df[df['Province'] == 'Western Cape']['TotalClaims']


# Perform the ANOVA test
f_stat, p_value_anova = stats.f_oneway(eastern_cape_claims, free_state_claims, gauteng_claims, kwazulu_natal_claims, limpopo_claims, mpumalanga_claims, north_west_claims, western_cape_claims)

print("--- ANOVA Results ---")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value_anova:.4f}")
print(f"Alpha level: 0.05")
print()

# Interpret the result
if p_value_anova < 0.05:
    print("✅ Result: Reject the null hypothesis.")
    print("   There is a significant difference in total claims between provinces.")
else:
    print("❌ Result: Fail to reject the null hypothesis.")
    print("   No significant difference in total claims between provinces.")

print(f"\nDescriptive Statistics by Province:")
province_stats = df.groupby('Province')['TotalClaims'].agg(['mean', 'std', 'count']).round(2)
print(province_stats)
```

#### H₀:There are no risk differences between zip codes 


```python
# Get list of provinces
print(df.groupby('PostalCode')['TotalClaims'].mean().round(2))
```

```python
from scipy import stats
import pandas as pd

# Make sure PostalCode is treated as categorical
df['PostalCode'] = df['PostalCode'].astype(str)

# Group claims by postal code
groups = [group["TotalClaims"].values for _, group in df.groupby("PostalCode")]

# Perform the ANOVA test
f_stat, p_value_anova = stats.f_oneway(*groups)

print("--- ANOVA Results ---")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value_anova:.4f}")
print(f"Alpha level: 0.05\n")

# Interpret the result
if p_value_anova < 0.05:
    print("✅ Result: Reject the null hypothesis.")
    print("   There is a significant difference in total claims between postal codes.")
else:
    print("❌ Result: Fail to reject the null hypothesis.")
    print("   No significant difference in total claims between postal codes.")

print("\nDescriptive Statistics by Postal Code:")
postal_stats = df.groupby("PostalCode")["TotalClaims"].agg(["mean", "std", "count"]).round(2)
print(postal_stats)
```

#### H₀:There are no significant margin (profit) difference between zip codes 

```python
from scipy import stats
import pandas as pd

# Make sure PostalCode is treated as categorical
df['PostalCode'] = df['PostalCode'].astype(str)

# Group claims by postal code
groups = [group["Margin"].values for _, group in df.groupby("PostalCode")]

# Perform the ANOVA test
f_stat, p_value_anova = stats.f_oneway(*groups)

print("--- ANOVA Results ---")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value_anova:.4f}")
print(f"Alpha level: 0.05\n")

# Interpret the result
if p_value_anova < 0.05:
    print("✅ Result: Reject the null hypothesis.")
    print("   There is a significant difference in margin between postal codes.")
else:
    print("❌ Result: Fail to reject the null hypothesis.")
    print("   No significant difference in margin between postal codes.")

print("\nDescriptive Statistics by Postal Code:")
postal_stats = df.groupby("PostalCode")["Margin"].agg(["mean", "std", "count"]).round(2)
print(postal_stats)
```

### Chi-Squared Test

#### - H₀:There are not significant risk difference between Women and Men

```python
# Create a contingency table (crosstab)
contingency_table = pd.crosstab(df['Gender'], df['HasClaim'])
print("Contingency Table:")
print(contingency_table)
print()

# Perform the Chi-Squared test
chi2_stat, p_value_chi2, dof, expected = stats.chi2_contingency(contingency_table)

print("--- Chi-Squared Test Results ---")
print(f"Chi-Squared Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value_chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Alpha level: 0.05")
print()

# Interpret the result
if p_value_chi2 < 0.05:
    print("✅ Result: Reject the null hypothesis.")
    print("   There is a significant association between gender and having claims.")
else:
    print("❌ Result: Fail to reject the null hypothesis.")
    print("   There is no significant association between the variables.")

# Show the expected frequencies
print(f"\nExpected Frequencies:")
expected_df = pd.DataFrame(expected, 
                          index=contingency_table.index, 
                          columns=contingency_table.columns)
print(expected_df.round(2))
```
