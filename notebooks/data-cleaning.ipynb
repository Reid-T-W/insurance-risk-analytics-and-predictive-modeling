```python
# Drop columns with missing values greater than 5%
df = df.drop(['Bank', 'Citizenship', 'CustomValueEstimate', 'NewVehicle', 'WrittenOff', 'Rebuilt', 'Converted', 'CrossBorder', 'NumberOfVehiclesInFleet'], axis = 1)
```

```python
# no of records in df
print("No of records in data frame")
print(len(df))
```

```python
# Checking count of records with missing values less than 5%
null_percent = df.isnull().mean() * 100
cols_with_nulls = null_percent[null_percent > 0]
display(cols_with_nulls)
```

```python
# Dropping rows with any type of missing values
df_cleaned = df.dropna()
```

```python
# Checking count of records with missing values less than 5%
print("no of missing values in each column after cleaning")
null_percent = df_cleaned.isnull().mean() * 100
cols_with_nulls = null_percent[null_percent > 0]
display(cols_with_nulls)

print("No of Records in cleaned dataframe")
print(f"No of Records in cleaned dataframe {len(df_cleaned)}")
```

```python
# Clean wrong records
# Some capital outstanding records have commas as decimals
# Replace them with dots
invalid_rows = df[df['CapitalOutstanding'].astype(str).str.contains(',', regex=False)]
print("No of invalid rows (containing commas) before correction: ", len(invalid_rows))

df['CapitalOutstanding'] = df['CapitalOutstanding'].astype(str).str.replace(',', '.')
df['CapitalOutstanding'] = pd.to_numeric(df['CapitalOutstanding'], errors='coerce')
invalid_rows = df[df['CapitalOutstanding'].astype(str).str.contains(',', regex=False)]

print("No of invalid rows (containing commas) after correction: ", len(invalid_rows['CapitalOutstanding']))
```

```python
# Clean wrong records
# Some total premium values have -ve values, which should not be
length = len(df[df["TotalPremium"] < 0])
print(f"total values less than 0 {length} before fix")

df = df[df['TotalPremium'] >= 0]

length = len(df[df["TotalPremium"] < 0])
print(f"total values less than 0 {length} after fix")
```
